## Numpy Arrays

NumPy is a powerful Python library for numerical computations. It stands for "Numerical Python" and provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

**Here are some key features and functionalities of NumPy:**

#### Ndarray: 
NumPy's main object is the ndarray (n-dimensional array). It is a homogeneous collection of elements with a fixed size in memory and supports fast element-wise operations, slicing, indexing, and reshaping. The ndarray is the foundation for most numerical computations in NumPy.

#### Mathematical Functions: 
NumPy provides a wide range of mathematical functions that can be applied element-wise to arrays. These include basic arithmetic operations, trigonometric functions, logarithmic functions, exponential functions, and more. These functions are optimized for performance and can operate on entire arrays without the need for explicit loops.

#### Array Manipulation: 
NumPy offers several functions to manipulate arrays, such as reshaping, slicing, joining, splitting, and transposing. These operations allow you to extract specific elements, subarrays, or rearrange the shape of arrays according to your needs.

#### Broadcasting: 
NumPy's broadcasting feature allows arrays of different shapes to be operated together. It simplifies computations by automatically aligning arrays with compatible shapes and performing element-wise operations.

#### Linear Algebra: 
NumPy provides a comprehensive suite of linear algebra functions. You can perform operations like matrix multiplication, determinant calculation, inverse calculation, eigenvalue and eigenvector computation, solving linear systems of equations, and more.

#### Random Number Generation: 
NumPy includes a submodule called random that provides functions for generating random numbers from various probability distributions. These functions are useful for simulations, statistical analysis, and generating random samples for testing.

#### Integration with other Libraries: 
NumPy is often used in conjunction with other scientific computing libraries, such as SciPy, pandas, and Matplotlib. These libraries build on top of NumPy to provide additional functionality for scientific computing, data analysis, and visualization.

### np.arange 
The np.arange function is a NumPy function that creates a one-dimensional array with evenly spaced values within a specified range. The syntax for np.arange is as follows:

```python
np.arange(0,10) # np.arange(start,stop,step) 
```

```python
np.arange(0,11,2) # All even number in a array
```

```python
np.arange(1,10,2) # All odd even number in a array
```
