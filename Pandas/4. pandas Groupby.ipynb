To put it simply group-by allows you to group toghther rows based off a column and then perform some sort of aggegate function on them

For instance in this example here we have three partitions of ID's one,two and three and then we have several values for them
We can Group-by the id's column and aggregate them using some sort of aggregate function ,

In every Function is basically just a fancy term for any function that takes many values and then split out or outputs a single value such as taking the sum of a bunch of values
and putting the result or taking the averge or standard deviation etc.

These are all aggregate function and that's all a group by method does it allows you to choose a column to group by 
and  it gathers all those rows toghther based off that column value and then you can perform some sort of aggregate function on it

```python
# Importing the modules 
import numpy as np
import pandas as pd
```

```python
data = { "company" : ["TCS","TCS","TCS","Goldman","Goldman","Goldman","4crisk","4crisk","4crisk"],
        "Person" : ["Raj","Rahul","Vivek","Pratik","Shubham","Vaishnavi","Ramendra","Deepanshu","Rishabh"],
        "Sales" : [1000,900,800,700,600,500,400,300,200]}
        
```

```python
data
```

```python
df = pd.DataFrame(data) # Converting dictionary to DataFrame
df
```

* 1st Columns tell us the company names TCS , Goldman , 4crisk

* 2nd column tell us the bunch of people working for that companies

* 3rd column tell us the sales genrated by the Peoples.

Now what we can do is use that group by method to group rows toghther based off a column name 

for instance let's go ahead and group by the company i'm going to go on and say ZF and then you say group

```python
bycomp = df.groupby("company")
bycomp
```

```python
bycomp.mean()
```

* The Reason it can't do this for the person column becaause these are strings and there's is no way can just make the mean of a string and that means just automatically ignores anything that's 
* Pandas automatically column which is Non-Numeric column.
* Likewise we can go ahead and perfrom other aggregate frunction such as sum of Everything

```python
bycomp.sum()
```

* You can even do standard deviation the vestey the etc and you should notice that what i get back an actual DataFrame with an index called company and the column value of sales
so we can also stuffs such as index offer so i could say by come up

```python
bycomp.std()
```

```python
bycomp.sum()
```

```python
bycomp.sum().loc["TCS"]
```

That gives us the sales of the TCS with all the sales value in it .

* Probably we will not do this by giving it to object like bycomp 
* let's what has to be actually be done:

```python
df.groupby("company").sum().loc["TCS"]
```

* Its all the thing in one line now.

#### Some other useful aggregate function that you may find yourself doing are thing's such as count
* Which count's the number of instances or column

```python
df.groupby("company").count()
```

* In this case it was able to return the person column because it's able to count how many instance's of of person occur at that column and that make's sense.


#### We can also do things such as Max and Min

```python
df.groupby('company').max()
```

* The max is the latest toward the end of the alphabet and if you call Min  that's the earliest in the

```python
df.groupby('company').min()
```

* *You should not use aggreate function with string column you should really use it with Numeric COlumns*
