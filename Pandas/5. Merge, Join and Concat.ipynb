## Concatenation :

Concatenation basically glues toghther Frames . Keep In mind that dimensions should match along the axis you are concatenation on. you can use pd.concat and pass in a list of Dataframes to concatenate toghther

```python
pd.concat([df1,df2,df3])
```

* By default the axis = 0 that means it will join it by rows
* We see here that succesfully concatenated or basically glued toghther the dataframes.
* We can see here A,B,C,D as columns and Index from 0-11

- If you want to Concatenate Along the Column YOu can use the axis = 1 

```python
pd.concat([df1,df2,df3],axis=1)
```

* And this is an example of what happens when you see Pd.concat on these three DataFrames along x-axis = 1 
* Notice here we have a bunch of missing values and that's because these dataframes didn't have values for all the indices that you wanted to concatenate on.
* For Example In Df1 A,B,C,D we had indices 0-3 we got the values but then we have indices 4-7 in df2 that's why it left the NaN values if we have the same Index values in Every Df we get the Result we Wanted Let's see that

## Merging:
The Merge function allows you to merge DataFrames toghther using a logic similar as merging SQL Tables Toghther , For example

```python
pd.merge( left , right , how="inner" , on="key" )
```

* The important thing here is *On Keyword* , In which basis we are merging the dataframe Left and Right .
* The important thing It is bydefault set to How = Inner we can also change this thing is ccalled Hyperparameter Tuning

* But basically what's happenning here is we'are going to combine or merge these dataframes toghther on these columns and that way instead of concatenating them we would have two key column we can merge so that we merge them or join them on the same key that they shares

* So, the logic here is when we are using the pd.merge instead of gluing the dataframes toghther you are going to where the value's matching up the columns

#### Let's see some Complicated Example's

```python
left = pd.DataFrame({ "key1" : ["K0","K0","K1","K2"],
                     "key2" : ["K0","K1","K0","K1"],
                     "A": [ "A0","A1","A2","A3"],
                     "B": [ "B0","B1","B2","B3"]})

right = pd.DataFrame({ "key1" : ["K0","K1","K1","K2"],
                     "key2" : ["K0","K0","K0","K0"],
                     "C": [ "C0","C1","C2","C3"],
                     "D": [ "D0","D2","D2","D3"]})
```

```python
pd.merge(left,right,on=["key1","key2"])
```

* it only add's the Inside matching value to give's us the dataframe

```python
pd.merge(left,right,how="outer",on=["key1","key2"])
```

- It show's up all the values and which doesn't add up Gives NaN values at that place

```python
pd.merge(left,right,how="left",on=["key1","key2"])
```

It start's on the basis of our left DataFrame and give's us the dataframe

## Joining:

joining is a convininent method for combining the columns of two potentially differntly-indexed DataFrames into a single result DataFrame.

```python
#Let's make the Dataframes
left = pd.DataFrame({"A": [ "A0","A1","A2"],
                     "B": [ "B0","B1","B2"]
                    },
                    index =  ["K0","K1","K2"])

right = pd.DataFrame({"C": [ "C0","C1","C2"],
                     "D": [ "D0","D2","D2"]
                     },index =  ["K0","K2","k3"])
```

* You could essentially think of this as the same thing except the key's you want to join on are actually on your index instead of a column

```python
left.join(right)
```

* Based off of the index key's if you want to merge these two dataframes based off of one of their column we would use merge
* But, Join method functionality comes off of one of the dataframes. You say that join and then pass other dataframe and just like merge you can specify inner , left , right , Outer, etc...
