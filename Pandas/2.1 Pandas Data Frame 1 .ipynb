* Pandas is a popular open-source Python library widely used for data manipulation and analysis. It provides powerful data structures like DataFrames and Series, which make working with structured data more intuitive and efficient. Here, I'll focus on explaining the Pandas DataFrame.

#### DataFrame:
A DataFrame is a 2-dimensional labeled data structure in Pandas, similar to a spreadsheet or a SQL table. It consists of rows and columns, where each column can have a different data type (e.g., numeric, string, datetime) and is identified by a column label. The rows are identified by an index, which provides a unique identifier for each row.

#### Removing the Column in the DataFrame.

```python
# Very Importanyt:

df.drop("New")
```

```python
#IF you just do it it gives us the Error
# What we have to is
df.drop("New",axis=1,inplace=True) 

# Here we have to drop the whole column
# Axis = 1 means Column
# Inplace mean  inplace : bool, default False
#    If False, return a copy. Otherwise, do operation
#    inplace and return None.
#errors : {'ignore', 'raise'}, default 'raise'
# I will explain in simple words
```

#### Inplace 
*  Inplace means we dont have to save our df again it will do it explicitly
*  If we dont use inplace we have to write it this way
*  df = df.drop("New",axis = 1)
*  If we dont use inplace the drop function will not work and we see the same column in the dataFrame again.

**If we want remove the whole Row we can use Axis = 0.** 

```python
df = df.drop("E",axis = 0) # by default the axis=0 for showing you i am writing it.
df
```
