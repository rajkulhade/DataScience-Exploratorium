

# "Missing data"
* A lot of times when we are using Pandas to read in data, If we have the Missing Ponts what's going to Happen is Pandas will automatically fill the Missing Point Null or NaN Values.
* We will see Many way's to fill in Missing data

```python
# Importing the Modules:

import numpy as np
import pandas as pd
```

```python
# This time we will create a DataFrame Using the Dictionary
# This dictionary will have three keys and it is going to be A,B,C and these are goig to be columns in the dataframe

d = {"A":[1,2,np.nan],"B":[5,np.nan,np.nan],"C":[1,2,3]}
```

```python
df = pd.DataFrame(d)
df
```

* Here we can see row 0 has no Missing values as well as Col C
* Remember in Dictionary we make a DataFrame Column wise
* We have 2 Missing values in Column B
* And 1 Missing Value in Column A 

## 1) Dropna()

* A Lot of time we just want to drop Missing values from the DataSet, Especially when it's just a few Missing values 
* It's a very Powerful Techniques Guy's , Be Very Careful it will drop the whole rows and reduce the size of the dataset
* It is recommended to use Dropna() , if Missing value's percentage is less than 5%.
* But i will recommened you to use the Data Imputation Techniques which we will see in Future
* Sometimes if missing values are less than 5% also it is not recommended to drop 
* **This is whole DOMAIN SPECIFIC**.

```python
df.dropna()
```

* Pandas will drop any row with any single NaN values contain on a Series


 
* By default the dropna is set to axis = 0 which means rows.
* We can also set to axis = 1

```python
df.dropna(axis = 1)
```

* This is going to Remove the Whole Column if a Single NaN value also present in it.
* It is Recomeded row Wise only at axis = 0\
* But, we can also go and Specify the *Threshhold* 
* if we go and check this out



```python
df.dropna()
```

* While using the dropna() it removed all the rows with a single NaN values also
* When using the Thresh with it, it will require specified Non-NaN to be in a list not to get dropped .
* Let's see it

```python
df.dropna(thresh = 2)
```

```python
df
```

* In row 2 there was only 1 Non-NaN values was there so it dropped that off
* In row 1 there is atleast two Non-NaN value so it left it off
* Till Now we have focused on removing the NaN Values Now lets see ways of Filling it.

# Fillna()

```python
# Let's took at the view at the DataFrame again
df
```



* In Fillna() we have the couple of argument's here
* See, we have the Value argument

```python
df.fillna(value="FILL VALUE")
```

* As we can have a look , We can fill any value we want using the value keyword)

* But many time's we want to fill the mean of the Column to the Missing value's,

```python
# Let's took at the column A
df["A"]
```

```python
# We want to fill the NaN value as Mean of the Column which is 1.5

df["A"].fillna(value = df["A"].mean())
```
