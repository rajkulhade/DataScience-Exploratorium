## Important Point:
* What makes Pandas Series different from NumPy Array is **Pandas Series holds a variety of Object types**
* Pandas Series holds almost hold every Data types

```python
pd.Series(data=labels) # It also holds String type
```

```python
pd.Series(data = [sum,print,len])
```

* **It also holds Built-in Function**.
* You can never do this in NumPy Array.
* This means How flexible is Pandas Series as far as holding different object types.
