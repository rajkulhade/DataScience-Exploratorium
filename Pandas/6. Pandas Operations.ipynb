### 4. Unique():
The nunique() function in the Pandas library is used to count the number of unique values in a Series or DataFrame column. It returns the count of distinct values present in the specified data.

```python
df["col2"].nunique()
```

### 5. value_counts():
 The value_counts() function is used to count the occurrences of unique values in a Series. It returns a new Series where the index consists of unique values from the original Series, and the values represent the counts of each unique value

In the dataframe above we can see:

```python
df["col2"].value_counts()
```

### 6. Apply Method

The apply function is going to be One of the most powerful tools in your hand when using Pandas.

The apply() function in the Pandas library is used to apply a given function along the axis of a DataFrame or Series. It's a powerful tool for transforming data in a flexible way. You can use it to apply a custom function, a built-in function, or even a lambda function to each element, row, or column of your data.

Let's see how it work's...

```python
# Imagine we have the function times2 which take in a single argument
def times2(x):
    return x*2

# Just a simple function
```

```python
# Now you know that you can already grab for instance a column and
# calls a Built-in Function

df["col1"].sum() # It returns us the sum of the columns..
```

 Now if we want a customized Function such as times2
 
 pandas has the ability to do that

```python
df["col1"].apply(times2)
```

 It will Broadcast each function in the column

#### You can also apply the Built-in function inside the apply

```python
# If you want the length of all the string in the col3
df
```

```python
df["col3"].apply(len)
```

```python
# You can also use lambda function inside the apply 

df["col2"].apply(lambda x: x*2)
```

We were able to multiply each value in that column by two and we were able to do this without having to write the Entire Function

### 12. Pivot Table

In pandas, a pivot table is a way to summarize and aggregate data from a DataFrame by grouping and rearranging it into a new table format. Pivot tables are especially useful for performing data analysis and generating summary reports. The pivot_table() function in pandas is used to create pivot tables. It allows you to specify which columns to use as rows, columns, and values, and you can apply aggregation functions to the values.

```python
# Let's make the New DataFrame to understand this:

data = {"A":["foo","foo","foo","bar","bar","bar",],
        "B":["one","one","two","two","one","one"],
        "C":["x","y","x","y","x","y",],
        "D":[1,3,2,5,4,1]}

df = pd.DataFrame(data)
df
```

pivot_table(data, values=None, index=None, columns=None, aggfunc='mean')

Pivot table has these many things and we have to idenify this.

```python
df.pivot_table(values="D",index=["A","B"],columns=["C"])


# The values we want in our table is D column values
# That's we want our DataFrame to be made up of
# That's why we say values="D"

# And then we will set our index to be equal to the A column
# and the B column that basically gonna make A & B
# into a Multilevel Index
# so we are going to pass list then since it's more than 
# One Column index A and B 

# And then we want the Actual column to be defined by
# the C column
```
